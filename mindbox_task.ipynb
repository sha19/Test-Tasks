```python
#ПЕРВАЯ ФУНКЦИЯ

# 1. Функция, которая подсчитывает число покупателей, попадающих в каждую группу,
#   если нумерация ID сквозная и начинается с 0
# На вход функция получает целое число n_customers (количество клиентов)

def function1(n_customers):
    groups = []
    for ID in range(0, n_customers + 1):
        N = ID
        sum = 0
        while N != 0:
            last = N % 10 #mod 10 всегда даёт нам последнюю цифру от числа N(в качестве остатка от деления числа N на 10)
            sum = sum + last
            #sum += last    альтернативная запись
            N = N // 10 #после того, как очередная последняя цифра занесена в сумму, её необходимо отсечь
                        # с помощью операции div (//)
        print('Номер группы, определяемый суммой цифр, для клиента с ID=%s' % ID, 'равен', sum)
        groups.append(sum)
    print('\n', groups, '\n')    

    #словарь из пар номеров групп и количеством входящих в группу клиентов
    from collections import Counter
    c = Counter(groups)
    print (c)
    
    #вывод элементов словаря
    print ('\nРаспределение клиентов в количестве %s' % n_customers, 'по группам:')
    for key, value in c.items():
        print ('Группа', key,':',value,'кл')
```

```python
function1(123)
```

```python
#ВТОРАЯ ФУНКЦИЯ

# 2. Функция, аналогичная первой, если ID начинается с произвольного числа
# На вход функция получает целые числа: n_customers (количество клиентов) и
#                                       n_first_id (первый ID в последовательности)

def function2(n_first_id, n_customers):
    groups = []
    for ID in range(n_first_id, n_first_id + n_customers + 1):
        N = ID
        sum = 0
        while N != 0:
            last = N % 10 #mod 10 всегда даёт нам последнюю цифру от числа N(в качестве остатка от деления числа N на 10)
            sum = sum + last
            #sum += last    альтернативная запись
            N = N // 10 #после того, как очередная последняя цифра занесена в сумму, её необходимо отсечь
                        # с помощью операции div (//)
        print('Номер группы, определяемый суммой цифр, для клиента с ID=%s' % ID, 'равен', sum)
        groups.append(sum)
    print('\n', groups, '\n')    

    #словарь из пар номеров групп и количеством входящих в группу клиентов
    from collections import Counter
    c = Counter(groups)
    print (c)
    
    #вывод элементов словаря
    print ('\nРаспределение клиентов в количестве %s' % n_customers,\
           'по группам ( когда first_id =', n_first_id, ', last_id =', n_first_id + n_customers,'):')
    for key, value in c.items():
        print ('Группа', key,':',value,'кл')
```

```python
function2(100, 123)
```
